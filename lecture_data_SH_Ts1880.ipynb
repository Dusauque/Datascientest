```python
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']


selected_years = df[df['Year'] % 10 == 0]


plt.figure(figsize=(12, 8))
colors = plt.cm.viridis(np.linspace(0, 1, len(selected_years)))

for (index, row), color in zip(selected_years.iterrows(), colors):
    year = int(row['Year'])
    temps = row[months].values
    plt.plot(months, temps, marker='o', linestyle='-', label=str(year), color=color)


plt.xlabel('Mois')
plt.ylabel('Anomalie de TempÃ©rature (Â°C)')
plt.title('Variation de la TempÃ©rature (1880 - 2025)')
plt.legend(title='AnnÃ©e', bbox_to_anchor=(1.05, 1), loc='upper left', fontsize=8)
plt.grid(True)


plt.show()
```
