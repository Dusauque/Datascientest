```python
df = pd.read_csv("SH.Ts+dSST.csv", sep=",", skiprows=1)
df
```

```python
df_na = (df.isna().mean()*100).round(2)
df_na
```

```python
df.plot(x='Year', y='Jan', marker='o')
```

```python
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']



df[months] = df[months].apply(pd.to_numeric)

df["Mean_Year"] = df[months].mean(axis=1)


plt.figure(figsize=(12, 6))
plt.plot(df["Year"], df["Mean_Year"], marker='o', linestyle='-', color='g', label="Température Moyenne Annuelle")


plt.xlabel('Année')
plt.ylabel('Anomalie de Température (°C)')
plt.title('Évolution de la Température Moyenne Annuelle (1880-2025)')
plt.legend()
plt.grid(True)

plt.show()
```

```python
df_airs = pd.read_csv("SH.Ts+dSST_AIRS.csv", sep=",", skiprows=1)
df_airs
```

```python
df_airsv6 = df_airs.iloc[1:23]

df_airsv7 = df_airs.iloc[27:49]

df_airsGHC = df_airs.iloc[53:75]
df_airsGHC
```

```python
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']

df_airsv6[months] = df_airsv6[months].apply(pd.to_numeric)
df_airsv7[months] = df_airsv7[months].apply(pd.to_numeric)
df_airsGHC[months] = df_airsGHC[months].apply(pd.to_numeric)


df_airsv6["Mean"] = df_airsv6[months].mean(axis=1)
df_airsv7["Mean"] = df_airsv7[months].mean(axis=1)
df_airsGHC["Mean"] = df_airsGHC[months].mean(axis=1)

plt.figure(figsize=(12, 8))
plt.plot(df_airsv6['Year'], df_airsv6['Mean'], label='AIRS V6 - Moyenne Annuelle', color='blue', marker='s')
plt.plot(df_airsv7['Year'], df_airsv7['Mean'], label='AIRS v7 - Moyenne Annuelle', color='green', marker='x')
plt.plot(df_airsGHC['Year'], df_airsGHC['Mean'], label='GHCNv4 - Moyenne Annuelle', color='red', marker='s')

plt.title('Carte Thermique des Anomalies de Température Mensuelles')
plt.xlabel('Année')
plt.ylabel('Mois')
plt.grid(True)
plt.legend()


plt.show()
```
