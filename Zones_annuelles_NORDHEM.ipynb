```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
df = pd.read_csv('ZonAnn.Ts+dSST vs 51 81.csv', sep=',', header = 0)
df.head()
```

```python
df.info()
```

```python
# @title Year vs NHem

from matplotlib import pyplot as plt
import seaborn as sns
def _plot_series(series, series_name, series_index=0):
  palette = list(sns.palettes.mpl_palette('Dark2'))
  xs = series['Year']
  ys = series['NHem']

  plt.plot(xs, ys, label=series_name, color=palette[series_index % len(palette)])

fig, ax = plt.subplots(figsize=(10, 5.2), layout='constrained')
df_sorted = df.sort_values('Year', ascending=True)
_plot_series(df_sorted, '')
sns.despine(fig=fig, ax=ax)
plt.xlabel('Year')
_ = plt.ylabel('NHem')
```

```python
plt.figure(figsize=(12,5))  # Dfinir la taille du graphique
plt.plot(df["Year"], df["NHem"], linestyle='-', color='blue', label="Temprature Nord H")

# Ajout de titres et labels
plt.xlabel("Anne")
plt.ylabel("Anomalie de Temprature (C) vs Moyenne 1951-1980")
plt.title("volution de la Temprature Nord H")
plt.legend();
```

```python
plt.figure(figsize=(12, 5))
sns.regplot(x=df["Year"], y=df["NHem"], scatter_kws={"color": "blue", "s":10}, line_kws={"color": "red"})

# Ajout de titres et labels
plt.xlabel("Anne")
plt.ylabel("Anomalie de Temprature (C) vs Moyenne 1951-1980")
plt.title("volution de la Temprature Nord H avec Rgression Linaire");
```
